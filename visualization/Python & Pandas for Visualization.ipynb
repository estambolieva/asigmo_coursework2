```python
import pandas as pd
import seaborn as sns
%matplotlib inline# output the plots directly into the Jupyter Notebook
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'mass': [0.330, 4.87 , 5.97],
                   'radius': [2439.7, 6051.8, 6378.1]},
                  index=['Mercury', 'Venus', 'Earth'])
df.plot.pie(y='mass', figsize=(5, 5))
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv")
pd.set_option("display.max.columns", None) # ask Pandas not to hide columns in case there are many columns
df.head()
```

```python
df.plot(x="Rank", y=["P25th", "Median", "P75th"]) # omitted  kind='line' as this is the default value
```

```python
df.plot(x="Rank", y=["P25th", "Median", "P75th"], kind='area')
```

```python
df.plot(x="Rank", y="P75th")
```

```python
df.Median.plot(kind="bar")
```

```python
df.Median.plot(kind="area")
```

```python
ax_h = df.Median.plot(kind="hist")
df.Median.plot(kind='kde', ax=ax_h, secondary_y=True)
```

```python
# Violin plot
sns.violinplot(x=df.Median);
# visualizes distribution and its probability density
# the median is the WHITE DOT in the middle

# for the rest - read here https://datavizcatalogue.com/methods/violin_plot.html
```

```python
df.Median.plot(kind='box')
# draws a line for the median
# a box around the 25 and 75 percentiles
```

```python
df.plot(kind='box', subplots=True, layout=(3,7))# we can apply a box plot on the full datafram
```

```python
df.plot(kind='box', subplots=True, layout=(3,7)); # the semicolon at the end hides matplotlib's messages
```

```python
df.plot(kind='box', subplots=True, layout=(3,7), figsize=(50, 30)); # we can apply a box plot on the full datafram
```

```python
## Checking for correlation
df.plot(x="Median", y="Unemployment_rate", kind="scatter");
```

```python
df.plot(x="Median", y="Unemployment_rate", kind="hexbin"); # the deeper the colour, the higher the density
```

```python
sns.violinplot(x='Median', y='Unemployment_rate', data = df);
```

```python
## checking for correlations again
corr = df.corr()
corr
```

```python
corr.style.background_gradient(cmap='coolwarm')
```

```python
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
dataset = 'https://raw.githubusercontent.com/albertsl/datasets/master/popularidad/temporal.csv' # link to more info about the data https://towardsdatascience.com/complete-guide-to-data-visualization-with-python-2dd74df12b5e
df = pd.read_csv(dataset)
df['Mes'] = pd.to_datetime(df['Mes'])
df.head()
```

```python
format_dict = {'data science':'${0:,.2f}', 'Mes':'{:%m-%Y}', 'machine learning':'{:.2%}'}
df.head(10).style.format(format_dict)
```

```python
df.head(10).style.format(format_dict).highlight_max(color='darkgreen').highlight_min(color='#ff0000')
```

```python
df.head(10).style.format(format_dict).background_gradient(subset=['data science', 'machine learning'], cmap='BuGn')
```
